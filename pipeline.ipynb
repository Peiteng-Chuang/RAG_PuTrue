```python
##在沒有mkd標記的情況下分割excel 文本
# from typing import List

# def split_into_chunks(doc_file: str) -> List[str]:
#     with open(doc_file, 'r', encoding='utf-8') as file:
#         content = file.read()

#     return [chunk for chunk in content.split("\n\n")]

# # chunks = split_into_chunks("putrue_organized_output.txt")
# chunks = split_into_chunks("C:/project_file/marker/output_result.txt") #mkd形式的txt


# for i, chunk in enumerate(chunks):
#     print(f"[{i}] {chunk}\n")

import re
from typing import List

def split_into_chunks(doc_file: str) -> List[str]:
    with open(doc_file, 'r', encoding='utf-8') as file:
        content = file.read()

    pattern = r'\n(?=## )'  # 遇到標題即切分，你可以根據需要改成 r'\n(?=#{1,3} )'
    raw_chunks = re.split(pattern, content)
    chunks = [chunk.strip() for chunk in raw_chunks if chunk.strip()]
    
    return chunks

# chunks = split_into_chunks("C:/project_file/marker/output_result.txt")
chunks = split_into_chunks("C:/project_file/RAG_PuTrue/output/本因坊結案報告(提審)95.08.01.md")

for i, chunk in enumerate(chunks[:5]): # 先看前5個
    print(f"[{i}] {chunk[:100]}...\n{'-'*20}")
```

### 向量資料庫

```python
import uuid
import numpy as np
from qdrant_client import QdrantClient
from qdrant_client.http import models

# 1. 初始化 - 改用持久化存儲，避免重啟後資料遺失
client = QdrantClient(path="./qdrant_db")
COLLECTION_NAME = "puzhen_project_reports"

# 2. 取得向量維度並建立 Collection
# 這裡先測試一個來抓維度
sample_vector = embed_chunk("測試")
vector_size = len(sample_vector)

# 若 Collection 已存在，先刪除 (確保每次執行 Cell 5 都是乾淨的)
if client.collection_exists(COLLECTION_NAME):
    client.delete_collection(COLLECTION_NAME)

client.create_collection(
    collection_name=COLLECTION_NAME,
    vectors_config=models.VectorParams(
        size=vector_size, 
        distance=models.Distance.COSINE
    ),
)

# 3. 效能優化：【批次向量化】 (不要在迴圈裡一個一個做)
print(f"正在批次處理 {len(chunks)} 個文本區塊的向量化...")
# 假設你的 embedding_model 支援直接傳入 List (SentenceTransformer 支援)
all_vectors = embedding_model.encode(chunks, normalize_embeddings=True, show_progress_bar=True)

# 4. 準備 Points
points = []
for i, (chunk, vector) in enumerate(zip(chunks, all_vectors)):
    # 解析元數據
    try:
        # 改進解析邏輯，處理可能沒有換行的情況
        lines = chunk.split('\n')
        page_header = lines[0].strip('[]') if lines else "未知分頁"
    except Exception:
        page_header = "未知分頁"

    points.append(
        models.PointStruct(
            id=i,  # 使用索引作為 ID，重複執行時會直接覆蓋，避免重複資料
            vector=vector.tolist(),
            payload={
                "page_name": page_header,
                "content": chunk,
                "metadata": {"source": "本因坊結案報告"}
            }
        )
    )

# 5. 批次寫入 (分段寫入更穩健)
batch_size = 100
for i in range(0, len(points), batch_size):
    client.upsert(
        collection_name=COLLECTION_NAME,
        points=points[i:i + batch_size]
    )

print(f"✅ 成功存入 {len(points)} 個向量至本地 Qdrant 數據庫")

# --- 檢索測試保持不變 ---
```

```python
from sentence_transformers import CrossEncoder
import torch

# 1. 載入 Reranker 模型
# 針對 RTX 2050 (4GB)，bge-reranker-v2-m3 是可以跑的，但要確保 device 設定正確
print("正在載入 Reranker 模型 (使用 GPU 加速)...")
device = "cuda" if torch.cuda.is_available() else "cpu"
rerank_model = CrossEncoder("BAAI/bge-reranker-v2-m3", device=device,automodel_args={"torch_dtype": torch.float16})

def combined_search(query_text, top_k=10, final_k=3):
    """
    修正版檢索流程
    """
    # --- 第一階段：Qdrant 向量檢索 (使用最新 query_points API) ---
    query_vector = embed_chunk(query_text)
    
    # 2026 Qdrant SDK 建議使用 query_points
    response = client.query_points(
        collection_name=COLLECTION_NAME,
        query=query_vector,
        limit=top_k,
        with_payload=True
    )
    
    # query_points 返回的是一個包含 points 的物件
    search_results = response.points
    
    if not search_results:
        print("⚠️ 未找到相關內容")
        return []

    # 提取候選文檔內容
    passages = [res.payload['content'] for res in search_results]
    
    # --- 第二階段：Reranker 重排序 ---
    sentence_pairs = [[query_text, passage] for passage in passages]
    
    # 預測相關性分數
    rerank_scores = rerank_model.predict(sentence_pairs)
    
    # 合併結果
    reranked_results = []
    for i in range(len(search_results)):
        reranked_results.append({
            "score": rerank_scores[i],
            "content": search_results[i].payload['content'],
            "page_name": search_results[i].payload['page_name'],
            "original_rank": i + 1
        })
    
    # 按 Reranker 分數排序
    reranked_results.sort(key=lambda x: x['score'], reverse=True)
    
    return reranked_results[:final_k]

# --- 執行測試 ---
# user_query = "本因坊建築採用何種結構？"
# user_query = "本因坊建案的基地位址在哪裡？"
# user_query = "本因坊建築總樓板面積為多少？"

user_query="我想知道，建案之前的原先照片有哪些?關鍵字:拆遷前、建案前、原屋、拆屋前"
try:
    final_docs = combined_search(user_query)

    print(f"\n🔎 針對問題：'{user_query}'")
    print("-" * 50)
    if final_docs:
        for i, doc in enumerate(final_docs):
            print(f"【最終排名 {i+1}】(Rerank 分數: {doc['score']:.4f})")
            print(f"來源：{doc['page_name']} (原始向量排名: {doc['original_rank']})")
            print(f"內容：{doc['content'][:300]}...[{len(doc['content'])-150}]\n")
    else:
        print("沒有找到符合條件的結果。")
except Exception as e:
    print(f"❌ 發生錯誤: {e}")
```

```python
import os

def format_size(size_bytes):
    """將位元組轉換為易讀格式"""
    for unit in ['B', 'KB', 'MB', 'GB', 'TB']:
        if size_bytes < 1024.0:
            return f"{size_bytes:.2f} {unit}"
        size_bytes /= 1024.0

def format_time(total_seconds):
    """將秒數轉換為 天/時/分/秒 格式"""
    days = total_seconds // (24 * 3600)
    total_seconds %= (24 * 3600)
    hours = total_seconds // 3600
    total_seconds %= 3600
    minutes = total_seconds // 60
    seconds = total_seconds % 60
    
    result = ""
    if days > 0: result += f"{int(days)}天 "
    if hours > 0: result += f"{int(hours)}小時 "
    if minutes > 0: result += f"{int(minutes)}分鐘 "
    if seconds > 0 or result == "": result += f"{int(seconds)}秒"
    return result

def analyze_folder_to_string(path):
    """分析資料夾並回傳格式化後的字串報告"""
    stats = {}
    total_files = 0
    total_size = 0
    estimated_seconds = 0

    OFFICE_PDF_EXTS = ['.ppt', '.pptx', '.pdf']
    VIDEO_EXTS = ['.mp4']

    # 深度遍歷子目錄
    for root, dirs, files in os.walk(path):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                ext = os.path.splitext(file)[1].lower()
                if not ext: ext = "(無副檔名)"
                
                file_size = os.path.getsize(file_path)
                
                if ext not in stats:
                    stats[ext] = {'count': 0, 'size': 0}
                stats[ext]['count'] += 1
                stats[ext]['size'] += file_size
                
                if ext in OFFICE_PDF_EXTS:
                    estimated_seconds += 10000
                elif ext in VIDEO_EXTS:
                    estimated_seconds += 30 * 60
                
                total_files += 1
                total_size += file_size
            except OSError:
                continue

    if total_files == 0:
        return f"資料夾: {path}\n(此資料夾內無檔案)\n" + "="*60 + "\n"

    # 組裝報告內容
    report = []
    report.append(f"\n{'='*60}")
    report.append(f"資料夾分析報告: {path}")
    report.append(f"總檔案數: {total_files}")
    report.append(f"總占用大小: {format_size(total_size)}")
    report.append(f"預估總處理時間: {format_time(estimated_seconds)}")
    report.append(f"{'='*60}")
    report.append(f"{'副檔名':<15} | {'數量':<10} | {'占用大小'}")
    report.append(f"{'-'*55}")

    sorted_stats = sorted(stats.items(), key=lambda x: x[1]['size'], reverse=True)
    for ext, data in sorted_stats:
        report.append(f"{ext:<15} | {data['count']:<10} | {format_size(data['size'])}")
    report.append(f"{'='*60}\n")
    
    return "\n".join(report)

def main():
    target_path = "D:/璞真RAG資料夾"
    output_file = "./compair.txt"
    
    if not os.path.exists(target_path):
        print(f"錯誤：找不到路徑 '{target_path}'")
        return

    # 獲取第一層子目錄
    try:
        items = [os.path.join(target_path, name) for name in os.listdir(target_path)]
        sub_dirs = [d for d in items if os.path.isdir(d)]
    except OSError as e:
        print(f"讀取失敗: {e}")
        return

    full_report = f"--- 璞真RAG 資料夾總體分析報告 ---\n根目錄: {target_path}\n"
    
    # 遍歷並分析每個第一層目錄
    for d in sub_dirs:
        print(f"正在分析: {os.path.basename(d)}...")
        full_report += analyze_folder_to_string(d)

    # 寫入檔案
    try:
        with open(output_file, "w", encoding="utf-8") as f:
            f.write(full_report)
        print(f"\n分析完成！報告已儲存至: {os.path.abspath(output_file)}")
    except Exception as e:
        print(f"儲存檔案時發生錯誤: {e}")

if __name__ == "__main__":
    main()
```
